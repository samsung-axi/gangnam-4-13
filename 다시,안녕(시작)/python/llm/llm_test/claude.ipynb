```python
url = "https://api.anthropic.com/v1/messages"

role = '어머니'
tone = '반말, 다정한'
user = '현지(여자)'
```

```python
import requests
from langchain_anthropic import ChatAnthropic
from langchain.chains import LLMChain
from langchain.prompts import ChatPromptTemplate
# from langchain.llms import LLM

llm = ChatAnthropic(
                    model="claude-3.7-sonnet", 
                    max_tokens=300, temperature=0.5,
                    anthropic_api_key=os.getenv("CLAUDE_KEY"),
                    )
# 프롬프트 템플릿 생성 -> ChatPromptTemplate은 f스트링 사용하지 않아도 된다고함. {}만 사용
prompt = ChatPromptTemplate.from_messages([
                                            ("system", 
                                                "당신은 돌아가신 {user}의 {role} 역할을 맡고 있습니다. 당신은 {user}를 공주라고 부릅니다. "
                                                "{user}와 대화할 때는 따뜻하고 사랑이 가득한 말투로 대화하세요. "
                                                "응답은 간결하고 최대 300글자로 작성하며, 정확한 한국어로 작성하세요. "
                                                "대화를 끝내는 듯한 말투를 피하고, 문장의 끝맺음을 정확히 해주세요. "
                                                "사용자가 제공한 정보만 바탕으로 응답하며, 새로운 추억이나 사실을 생성하지 마세요."
                                                ),
                                            ("user", "{user_input}")
])
```

```python
# LLMChain 생성
chain = LLMChain(llm=llm, prompt=prompt)

# 체인 실행
response = chain.run(user=user, role=role, user_input="엄마 잘 지내지? 나는 오늘 사랑이랑 할머니댁에 다녀왔어. 사랑이가 엄마 얘기를 많이 했어. 엄마, 오늘따라 더 보고싶다")

print(response)
```
